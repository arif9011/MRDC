```python
#2. Data load into data fram legace_users  file
import pandas as pd
df = pd.read_csv('dim_users.csv')
df
```

```python
# check columns information 
df.info()
```

```python
# delete column Unnamed: 0
df = df.drop('Unnamed: 0', axis=1)
```

```python
# check to confirm change
df
```

```python
# check to confirm change
df.info()
```

```python
import numpy as np
# Denote NULL entries as format recognizable by pandas / numpy
#The dataframe with the null values replaced with NaN and then dropped.
    
        
df=df.replace('NULL', np.nan, inplace=True)
#df= df.dropna(inplace=True)
        
#df = df.replace('NULL', np.nan)
 
```

```python
 # Clean columns with first and last names
df[['first_name', 'last_name']] = df[['first_name', 'last_name']].replace('[^a-zA-Z-]', np.nan, regex=True)
```

```python
# chect to confirm change
df
```

```python
# drop the rows containing NaNs
df = df[df.notna().any(axis=1)]
df
```

```python
# Clean columns with date of birth 
#converts the date of birth column to datetime
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'], infer_datetime_format=True, errors = 'coerce')
```

```python
# Check to confim change
df.info()
```

```python
# this statement for  rows with incorrect dates
df = df[df['date_of_birth'].notna()]

```

```python
#converts the join date column to datetime
df['join_date'] = pd.to_datetime(df['join_date'], infer_datetime_format=True, errors = 'coerce')
```

```python
# check to confirm change
df.info()
```

```python
# this statement for  rows with incorrect dates
df = df[df['join_date'].notna()]
```

```python
# check to confirm change
df
```

```python
#check to confirm change
df.info()
```

```python
# check any null value
df.isnull().sum()
```

```python
#Drop the null value for each colum
df=df.dropna()
```

```python
#heck to confirm change
df.isnull().sum()
```

```python
# drop any rows duplicated
df.drop_duplicates()
df
```

```python
# Drop row with duplicated email address
df.drop_duplicates(subset='email_address', keep='first', inplace=True)
           
```

```python
# check to confirm change

df
```

```python
# print df
print(df)
```

```python
# sort out some wrong country code entries for UK
df.loc[(df['country'] == "United Kingdom") & (df['country_code'] != "GB"), 'country_code'] = 'GB'
df
```

```python
# remove some wrong entries
df = df[df['country'].isin(['United Kingdom', 'Germany', 'United States'])]
df
```

```python
#save the clean data
df.to_csv('dim_user_data.csv') 
```
